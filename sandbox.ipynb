```python
import json

from matplotlib import pyplot as plt 
import seaborn as sns 
import numpy as np  
import pandas as pd   
```

```python
# path = "output/dataflow_dialogues_stats/valid.count_dict.json"
# data = json.load(open(path))

filtered_path = "filtered-histograms.json"

filtered_data = json.load(open(filtered_path))


train_examples = filtered_data['datasetExamples']['train']


print(filtered_data.keys())
data = filtered_data['datasetHistograms']['train']

```

```python
sorted_data = sorted(data.items(), key=lambda x: x[1])

MIN_THRESH=1000
MAX_THRESH=2000

# make a dumb df 
data_tidy = []
labels = []
print(sorted_data[0], sorted_data[-1])
print(len(sorted_data))
for k,v in sorted_data:
    if v < MIN_THRESH or v > MAX_THRESH:
        continue

    labels.append(k)
    for i in range(v): 
        data_tidy.append(k)

        
```

```python



fig, ax = plt.subplots(figsize=(100,10))

sns.histplot(data_tidy, ax=ax, stat = "frequency", discrete = True)

ax.set_xticklabels(labels, rotation = 90)

plt.tight_layout()
plt.savefig("/Users/Elias/Desktop/freq.pdf")
```
