```python
import numpy as np
import json
import sys 
from glob import glob 

sys.path.insert(0, "../miso/commands/")
from pearson_aggregate import compute_pearson_score
```

```python
import pandas as pd
from collections import defaultdict

def paths_to_df(data_paths):
    data = []
    for path in data_paths:
        model_name = path.split("/")[-2]
        model_data = {"name": model_name}
        dev_avg_baseline_f1, dev_avg_f1, dev_avg_rho, __, __, __ = compute_pearson_score(path, False )

        model_data["baseline_f1"] = dev_avg_baseline_f1
        model_data["dev_f1"] = dev_avg_f1
        model_data["dev_pearson"] = dev_avg_rho
        data.append(model_data)


    df = pd.DataFrame(data)
    return df
```

```python

data_paths = glob("/exp/estengel/miso_res/new_interface/*/data.json")

df = paths_to_df(data_paths)
df
```
